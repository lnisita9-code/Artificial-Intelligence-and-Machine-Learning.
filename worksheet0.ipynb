```python
def convert_units(choice, option, value):
    """
    Converts units based on user choice.

    Parameters:
    choice (str): User’s conversion category
    option (str): Selected conversion option
    value (float): Number to be converted

    Returns:
    float: Converted value
    """

    # Length
    if choice == "1":
        if option == "1":
            return value * 3.28084
        elif option == "2":
            return value / 3.28084

    # Weight
    elif choice == "2":
        if option == "1":
            return value * 2.20462
        elif option == "2":
            return value / 2.20462

    # Volume
    elif choice == "3":
        if option == "1":
            return value * 0.264172
        elif option == "2":
            return value / 0.264172


print("Choose conversion type:")
print("1. Length")
print("2. Weight")
print("3. Volume")

choice = input("Enter your choice (1/2/3): ")

try:
    value = float(input("Enter the value to convert: "))

    if choice == "1":
        print("1. Meters to Feet")
        print("2. Feet to Meters")

    elif choice == "2":
        print("1. Kilograms to Pounds")
        print("2. Pounds to Kilograms")

    elif choice == "3":
        print("1. Liters to Gallons")
        print("2. Gallons to Liters")

    option = input("Choose option: ")

    result = convert_units(choice, option, value)

    print("Converted value:", result)

except ValueError:
    print("Error: Please enter a valid number.")
```

```python
def find_sum(numbers):
    """
    Calculates the sum of numbers in a list.

    Parameters:
    numbers (list): A list containing numbers.

    Returns:
    float: The total sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of numbers in a list.

    Parameters:
    numbers (list): List of values.

    Returns:
    float: Average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Finds the maximum number in a list.

    Parameters:
    numbers (list): List of numbers.

    Returns:
    float: Maximum value in the list.
    """
    return max(numbers)


def find_min(numbers):
    """
    Finds the minimum number in a list.

    Parameters:
    numbers (list): List of numbers.

    Returns:
    float: Minimum value in the list.
    """
    return min(numbers)


print("Choose an operation:")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

choice = input("Enter your choice (1/2/3/4): ")

try:
    user_input = input("Enter numbers separated by spaces: ")
    numbers = [float(num) for num in user_input.split()]

    if len(numbers) == 0:
        raise ValueError("List cannot be empty")

    if choice == "1":
        result = find_sum(numbers)
    elif choice == "2":
        result = find_average(numbers)
    elif choice == "3":
        result = find_max(numbers)
    elif choice == "4":
        result = find_min(numbers)
    else:
        print("Invalid operation")
        exit()

    print("Result:", result)

except ValueError:
    print("Error: Please enter valid numbers.")
```

4.3 Exercise on Nested List:

```python
#Task 1: Flatten a Nested List
def flatten(lst):
    flat_list = []
    for sublist in lst:
        flat_list.extend(sublist)
    return flat_list
nested = [[1, 2], [3, 4], [5]]
flat = flatten(nested)
print(flat)
```

```python
#Task 2: Accessing Nested List Elements
def access_nested_element(lst, indices):
    element = lst
    for index in indices:
        element = element[index]
    return element
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)
print(result)
```

```python
#Task 3: Sum of All Elements in a Nested List
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)
print(result)
```

```python
#Task 4: Remove Specific Element from a Nested List
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)
    return result
nested_list = [[1, 2], [3, 2], [4, 5]]
elem_to_remove = 2
new_list = remove_element(nested_list, elem_to_remove)
print(new_list)
```

```python
#Task 5: Find the Maximum Element in a Nested List
def find_max(lst):
    max_value = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)
    return max_value
nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)
print(result)
```

```python
#Task 6: Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count
nested_list = [[1, 2], [2, 3], [2, 4]]
elem_to_count = 2
result = count_occurrences(nested_list, elem_to_count)
print(result)
```

```python
#Task 7: Flatten a List of Lists of Lists
def deep_flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)
print(result)
```

```python
#Task 8: Nested List Average
def average_nested(lst):
    def flatten_and_sum(lst):
        total = 0
        count = 0
        for item in lst:
            if isinstance(item, list):
                sub_total, sub_count = flatten_and_sum(item)
                total += sub_total
                count += sub_count
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = flatten_and_sum(lst)
    return total_sum / total_count if total_count != 0 else 0
nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)
print(result)
```
